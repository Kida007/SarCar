# Defining a function to assign Time Group to each accident

```python
def time(time):
    if(time>0 and time<600):
        return 1
    elif(time>600 and time<1200):
        return 2
    elif(time>1200 and time<1800):
        return 3
    else:
        return 4
```

```python
df['Time']=df['Time'].astype(int)
```

# Assigning different Time Groups

```python
df['Time']=df['Time'].apply(time)
```
