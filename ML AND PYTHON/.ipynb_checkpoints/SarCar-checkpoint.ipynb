Using the **Calculate_Path_Vector** function we calculate a 15 seconds Path_Vector with a TimeStamp of a Test Vehicle and it Gets Updated Every 0.1 Seconds under 5g Network Conditions

Now Their Might be An **Error** Calculating the **future Position of the Car** because of the Complex **Curvature of the Road ** Which can be fixed By Tweaking our **Constant Bearing Angle ** Instantaneously with the Help of a **Artifical Neural Network **(The ANN Function) which gets trained and Improved Every Second reading Road Curvatures Like a Book

```python
# Originally Calculated in The User-Based Sarcar's Application
import math
import numpy as np



#Lets Take an Example Lat, Long , Bearing  , Angle , Speed . 
     
brng = 77.57                   #Bearing Angle in Degrees
speed = 20                           # speed in Km/h
road_name = "Great River Road"       # Road Name For Better Classification of Curvature
lat1 = math.radians(40.65818)    #Current lat point converted to radians
lon1 = math.radians(-74.259915)       #Current long point converted to radians
timestamp1 =1704408075


def Calculate_Path_Vector(lat1 , lon1 , road_name , speed , brng , timestart ) : 
    
    R = 6378.1                           #Radius of the Earth

    path_lat = [] 
    path_long = [] 
    time_Stamper= [] 
    
    lat2=lat1 
    lon2=lon1
    
    for i in range(1,300) :
        time_tweaker = 4.16e-05         # .15 sec Converted into Hrs 
        
        #Calculate Distance
        d = speed*time_tweaker*i       
        
        #Fixing Bearing Angle 
        brng = Ann(road_name ,lat2 , lon2 , brng)
        
        # Calculating Future Latitude Longitude
        lat2 = math.asin( math.sin(lat1)*math.cos(d/R) +
                         math.cos(lat1)*math.sin(d/R)*math.cos(brng))
        lon2 = lon1 + math.atan2(math.sin(brng)*math.sin(d/R)*math.cos(lat1),
                                 math.cos(d/R)-math.sin(lat1)*math.sin(lat2))
        
        # Appending Path Vector 
        
        time  = timestart + i*150 ;
        time_Stamper.append(time) ; 
        path_lat.append(math.degrees(lat2))
        path_long.append(math.degrees(lon2))
        
        

    return (path_lat , path_long , time_Stamper)
            

def Ann ( road_name ,lat, lon , brng , train=False) :
    
    if(train is False) :
        return brng
    
    else :
        X= [road_name ,lat, lon , brng]
        nn = MLPClassifier() 
        bearning = nn.predict(X)
        return bearing

Path_Vector = Calculate_Path_Vector(lat1 , lon1 , road_name , speed , brng ,timestamp1)
print(Path_Vector)
```

## 3) Collision Prediction Using Dimensional Analysis
Now as we Have Data of  Nearby Vehicles . we can **Dimensionally Analysis** the **Paths(Path-Vectors)** with **Time-stamps** of The **Test-Vehicle** with all of the Nearby-Vehicles . This Will Provide us with  a Rough Preidiction about Collision . Here We are Comparing The **Test-Vehicle** with Nearby Vehicle **Car-T2131** . The First 3D Dimesional Graph shows the The **Comparision** of path Vectors of both the Vehicles . Path Vectors on 2 axis with Time Stamp on 3rd Axis . Now Using This Graph we Can Predict a **Sub-Collision-Index** for Every Single Instance , which is nothing but the **Closest Vehicle Approach** with a bias of ** Time Difference**. 

We are Calculating this **Sub_Collision-Index** using the **cal_sub_collision_index** Function . **Lower** The sub  Collision Index  means **less distant approach ** between two vehicles . 

**Graph 2** shows the Sub Collision Index with respect to timeStamp and also Highlights  the **Minimum Distant** Approach

Now we have the **Minimun Sub Collision Index** and we can apply a final **Neural Net Layer ** to Convert it into 
**Final Collision Index**

##  4. ) The Final Classification of Collision  Using a Filtering Neural Network  

Finally , On the basis of **Sub-Collision Index** , **speed of both Vehicles** , ** Length of Both Vehicles** ,  we are classifying a Collision or not Collision . we will use a **Multi-Layer Perceptron Classifier ** which will be pretrained . it will classify a collision with *Binary Represesntation* i.e ** 0 means No Collision** and ** 1 means  Collision** .  The Function **FNN** will do this task. We are using **sklearns** for the mlpclassifier . Since dont have enough data for training currently , we will use **td=False**


```python
from sklearn.neural_network import MLPClassifier

def FNN(sub_col_ind , speed1 , speed2 , lenght1 , lenght2 , td=False) :
    
    if(td==True) :
        X= [sub_col_ind , speed1 , speed2 , lenght1 , lenght2]
        nn = MLPClassifier() 
        col_index = nn.predict(X)
        return col_index
    
    else :
        rms_speed = math.fabs(speed1+speed2)
        if(sub_col_ind<(lenght1+lenght2) and  rms_speed>90) : 
            return 1
        else :
            return 0 
        
Col_index  = FNN(0.2647682351901954 , 40 , 60  , 0.00155343 , 0.00155343 , 0.00124274)
print("Final Collision Index of TestCar with Car-T2131 :" , Col_index)
```
